```python
from nlp.cleaning_tfidf import clean_for_tfidf, compute_tfidf
import pandas as pd
df = pd.read_pickle(os.path.join(PROCESSED_DATA_DIR, "even_more_articles_with_embeddings.pkl"))

df["text_tfidf"] = df.apply(
    lambda row: clean_for_tfidf(row["content"], row["language"]),
    axis=1
)

X_tfidf, tfidf_vectorizer = compute_tfidf(df["text_tfidf"])
```

```python
import numpy as np
from nlp.clustering import find_optimal_k, fit_kmeans

EMBEDDING_MODEL = "miniLM_multilingual"
X_embeddings = np.vstack(
    df[f"embedding_{EMBEDDING_MODEL}"].values
)
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
def find_optimal_k2(embeddings, k_min=3, k_max=10):
    """
    Busca el K óptimo usando silhouette score
    """
    scores = {}

    for k in range(k_min, k_max + 1):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init="auto")
        labels = kmeans.fit_predict(embeddings)

        score = silhouette_score(embeddings, labels, metric="cosine")
        scores[k] = score

    best_k = max(scores, key=scores.get)
    return best_k, scores

optimal_k, scores = find_optimal_k2(X_embeddings, k_min = 4, k_max = 12)
optimal_k

labels, centroids = fit_kmeans(X_embeddings, k=optimal_k)
df["cluster"] = labels
```

```python
from nlp.interpretation import top_terms_per_cluster, name_clusters

def top_terms_per_cluster2(
    X_tfidf,
    clusters,
    vectorizer,
    top_n=10
):
    """
    Extrae los términos más representativos de cada cluster
    a partir de una matriz TF-IDF ya entrenada.
    """

    terms = np.array(vectorizer.get_feature_names_out())
    cluster_terms = {}

    for cluster_id in np.unique(clusters):
        cluster_mask = clusters == cluster_id
        cluster_tfidf = X_tfidf[cluster_mask]

        # media TF-IDF por término dentro del cluster
        mean_tfidf = cluster_tfidf.mean(axis=0).A1

        top_indices = mean_tfidf.argsort()[::-1][:top_n]
        cluster_terms[cluster_id] = terms[top_indices].tolist()

    return cluster_terms


cluster_terms = top_terms_per_cluster2(
    X_tfidf,
    df["cluster"].values,
    tfidf_vectorizer,
    top_n=10
)


cluster_names = name_clusters(cluster_terms)
df["cluster_name"] = df["cluster"].map(cluster_names)
```

```python
import numpy as np

embeddings = np.vstack(df["embedding_miniLM_multilingual"].values)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(embeddings)
```

```python
SIM_THRESHOLD = 0.95

duplicates = []

n = similarity_matrix.shape[0]

for i in range(n):
    for j in range(i + 1, n):
        if similarity_matrix[i, j] > SIM_THRESHOLD:
            duplicates.append((i, j, similarity_matrix[i, j]))
```

```python
def top_terms_per_cluster(tfidf_matrix, df, vectorizer, cluster_id, top_n=10):
    idx = df[df["cluster"] == cluster_id].index
    mean_tfidf = tfidf_matrix[idx].mean(axis=0)
    terms = vectorizer.get_feature_names_out()
    top_idx = np.argsort(mean_tfidf.A1)[::-1][:top_n]
    return [terms[i] for i in top_idx]

top_terms_per_cluster(X_tfidf, df, tfidf_vectorizer, "cluster", top_n=10)
```

```python
df["length_score"] = df["content"].str.len()
df["novelty_score"] = 1 - df.index.map(lambda i: similarity_matrix[i].mean())

df["final_score"] = (
    0.5 * df["novelty_score"] +
    0.3 * (df["length_score"] / df["length_score"].max()) +
    0.2 * df["source"].isin(["OpenAI", "AWS"]).astype(int)
)
```

## HTML

```python
AMC_PROFILES = {
    "Data & AI": "machine learning artificial intelligence data science models",
    "IT & Cloud": "cloud infrastructure devops aws azure automation",
    "Business": "strategy innovation digital transformation business",
    "Risk & Compliance": "regulation ai ethics governance risk compliance"
}

profile_embs = embedder.encode(list(AMC_PROFILES.values()))


def assign_area(article_emb, profile_embs, areas):
    sims = cosine_similarity([article_emb], profile_embs)[0]
    return areas[np.argmax(sims)]

areas = list(AMC_PROFILES.keys())
df["area"] = [
    assign_area(e, profile_embs, areas)
    for e in embeddings
]

def generate_newsletter(df_area, area_name):
    html = f"<h1>Newsletter – {area_name}</h1><ul>"
    for _, row in df_area.sort_values("final_score", ascending=False).head(10).iterrows():
        html += f"""
        <li>
            <b>{row['title']}</b><br>
            <i>{row['source']}</i><br>
            <a href="{row['url']}">Leer más</a>
        </li>
        """
    html += "</ul>"
    return html
```
